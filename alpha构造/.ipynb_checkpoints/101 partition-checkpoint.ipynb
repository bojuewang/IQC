```python
import random

def random_partition(n=101, k=4, seed=None):
    if seed is not None:
        random.seed(seed)
    
    arr = list(range(1, n+1))
    random.shuffle(arr)  # 随机排列
    
    # 分块（尽量均匀）
    sizes = [n // k] * k
    for i in range(n % k):
        sizes[i] += 1
    
    blocks = []
    idx = 0
    for s in sizes:
        blocks.append(arr[idx:idx+s])
        idx += s
    
    return blocks


# 使用
blocks = random_partition(101, 4, seed=42)

for i, b in enumerate(blocks):
    print(f"B{i+1} (size={len(b)}):", b)
```

```python
blocks[3].sort()   # 第4个块（index 从0开始）
print(blocks[3])
```

```python
# 针对排序后的 B4 进行 3 个一组的划分
B4_sorted = blocks[3]
B4_grouped = [B4_sorted[i:i + 3] for i in range(0, len(B4_sorted), 3)]

# 打印结果查看
for idx, group in enumerate(B4_grouped):
    print(f"Group {idx + 1}: {group}")
```
